```python
# Import required libraries and dependencies
import pandas as pd
import numpy as np
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv("Resources/crypto_market_data.csv",index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(width = 800, height = 400, rot = 90)
```

```python
#Renaming Column Names
df_market_data = df_market_data.rename(columns={'price_change_percentage_24h': 'change24h', 'price_change_percentage_7d': 'change7d','price_change_percentage_14d': 'change14d', 'price_change_percentage_30d': 'change30d', 'price_change_percentage_60d': 'change60d', 'price_change_percentage_200d': 'change200d', 'price_change_percentage_1y': 'change1y'})
```

```python
df_market_data.head(25)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
normalizeddata = StandardScaler().fit_transform(df_market_data[['change24h', 'change7d','change14d', 'change30d','change60d', 'change200d','change1y']])
```

```python
# Create a DataFrame with the scaled data
scaleddata = pd.DataFrame(normalizeddata, columns=['change24h', 'change7d','change14d', 'change30d','change60d', 'change200d','change1y'])

# Copy the crypto names from the original data
scaleddata['Crypto Names']=df_market_data.index


# Set the coinid column as index
scaleddata = scaleddata.set_index('Crypto Names')


# Display sample data
scaleddata.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(scaleddata)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbowcurve = {"k_original": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbowcurveddf = pd.DataFrame(elbowcurve)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
inertialinechart = elbowcurveddf.hvplot.line(x = "k_original", y = "inertia", title = "Elbow Curve", xticks = k)

inertialinechart
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer: The best value for k is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaleddata)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeanspredictions = model.predict(scaleddata)

# Print the resulting array of cluster values.
kmeanspredictions
```

```python
# Create a copy of the DataFrame
scaleddataprediction = scaleddata.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
scaleddataprediction['Prediction']=kmeanspredictions


# Display sample data
scaleddataprediction.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
datascatterplot = scaleddataprediction.hvplot.scatter(x = "change24h", y = "change7d", by = "Prediction", hover_cols = ['Crypto Names'])

datascatterplot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
marketpca = pca.fit_transform(scaleddata)

# View the first five rows of the DataFrame. 
marketpca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
explainedvariance = np.sum(pca.explained_variance_ratio_)
explainedvariancepercentage = explainedvariance * 100
print(f' The Total Explained Variance is:', explainedvariancepercentage)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.50%

```python
# Create a new DataFrame with the PCA data.
pcadf = pd.DataFrame(marketpca, columns = ["PCA1", "PCA2" , "PCA3"])

# Creating a DataFrame with the PCA data
pcadf["Crypto Names"] = df_market_data.index

# Copy the crypto names from the original data
pcadf["Crypto Names"] = df_market_data.index


# Set the coinid column as index
pcadf = pcadf.set_index('Crypto Names')

# Display sample data
pcadf.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
kpca = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia=[]


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in kpca:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(pcadf)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbowcurve2 = {'k_pca': k_pca , 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbowcurveddf2 = pd.DataFrame(elbowcurve2)
elbowcurveddf2.head(10)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
inertialinechart2 = elbowcurveddf2.hvplot.line(x = "k_pca", y = "inertia", title = "Elbow Curve", xticks = k)

inertialinechart2
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No it does not

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model2 = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model2.fit(pcadf)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusterspredict = model2.predict(pcadf)


# Print the resulting array of cluster values.
clusterspredict
```

```python
# Create a copy of the DataFrame with the PCA data
pcdpredictiondf = pcadf.copy()

# Add a new column to the DataFrame with the predicted clusters
pcdpredictiondf['Pca Prediction'] = clusterspredict


# Display sample data
pcdpredictiondf.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pcascatterplot = pcdpredictiondf.hvplot.scatter(x = "PCA1", y = "PCA2", by = "Pca Prediction", hover_cols = ['Crypto Names'])

pcascatterplot
```
